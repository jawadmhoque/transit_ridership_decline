```python
employment=pd.read_csv('../Factors and Ridership Data/Employment/employment.csv', encoding='utf-8')
employment
```

```python
employment=employment.dropna(axis=1)
```

```python
employment
```

```python
employment['Area_Code']=employment['Series ID'].apply(lambda x:x[:-2])
employment
```

```python
merged=pd.merge(df,employment,on='Area_Code')
merged
```

```python
unemployment_rate=merged[merged['Data Type']=='Unemployment Rate']
unemployment_rate
```

```python
unemployment_labor=merged[merged['Data Type']=='Unemployment ']
employment_labor=merged[merged['Data Type']=='Employment']
total_labor=merged[merged['Data Type']=='Total Labor Force']
```

```python
test=unemployment_rate.head()
```

```python
test2=pd.melt(test, id_vars=list(test)[0:8], value_vars=list(merged)[9:],var_name='Month-Year',
       value_name='Unemployment_Rate')
```

```python
test2['Month-Year']=test2['Month-Year'].apply(lambda x: x.replace("\\n",""))
test2
```

```python
import datetime
test2['DateTime']=pd.to_datetime(test2['Month-Year'])
test2
```

```python
df_unemployment=pd.melt(unemployment_rate, id_vars=list(unemployment_rate)[0:8], value_vars=list(unemployment_rate)[9:],var_name='Month-Year',
       value_name='Unemployment_Rate')
df_unemployment
```

```python
df_unemployment_labor=pd.melt(unemployment_labor, id_vars=list(unemployment_labor)[0:8], value_vars=list(unemployment_labor)[9:],var_name='Month-Year',
       value_name='Unemployment_Labor')
df_unemployment_labor
```

```python
df_employment_labor=pd.melt(employment_labor, id_vars=list(employment_labor)[0:8], value_vars=list(employment_labor)[9:],var_name='Month-Year',
       value_name='employment_Labor')
df_employment_labor
```

```python
df_total_labor=pd.melt(total_labor, id_vars=list(total_labor)[0:8], value_vars=list(total_labor)[9:],var_name='Month-Year',
       value_name='Total_Labor')
df_total_labor
```

```python
df_total_labor=df_total_labor.drop_duplicates(['Series ID','Month-Year'])
df_unemployment_labor=df_unemployment_labor.drop_duplicates(['Series ID','Month-Year'])
df_employment_labor=df_employment_labor.drop_duplicates(['Series ID','Month-Year'])
df_unemployment=df_unemployment.drop_duplicates(['Series ID','Month-Year'])
```
