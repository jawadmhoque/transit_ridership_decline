```python
import pandas as pd
import numpy as np
```

```python
### Read UPT and VRM data, from NTD March 2019
ntd_upt=pd.read_csv('../Factors and Ridership Data/Final Datasets/NTD_UPT.csv',encoding='utf-8')
#ntd_upt.head()
ntd_vrm=pd.read_csv('../Factors and Ridership Data/Final Datasets/NTD_VRM.csv',encoding='utf-8')
#ntd_vrm
```

```python
###Bus and Rail Specification

rail=['AG','CC','CR','HR','LR','IP','MO','RL','RP','RS','SR','VT','MG','YR']
bus=['MB','RB','CB','TB']
dr = ['DT','JT','VP','DR','PB']
other=['TR','FB']
def railbus(df):
    if df['Modes'] in rail:
        return 'Rail'
    elif df['Modes'] in bus:
        return 'Bus'
    elif df['Modes'] in dr:
        return 'DemandResponse'
    else:
        return "Other"
```

```python
###Convert to long format 

df_upt=pd.melt(ntd_upt, id_vars=list(ntd_upt)[0:9], value_vars=list(ntd_upt)[9:],var_name='MonthYear',
       value_name='UPT')
df_vrm=pd.melt(ntd_vrm, id_vars=list(ntd_vrm)[0:9], value_vars=list(ntd_vrm)[9:],var_name='MonthYear',
       value_name='VRM')
```

```python
df_vrm.to_csv('../Factors and Ridership Data/Tests/df_vrm.csv')
```

```python
### combined dataframe containing UPT and VRM values for each mode and each TOS at each Month

merged_ntd=pd.merge(df_upt,df_vrm,on=['5 digit NTD ID',
 '4 digit NTD ID',
 'Agency',
 'Active',
 'Reporter Type',
 'UZA',
 'UZA Name',
 'Modes',
 'TOS',
 'MonthYear'],how='inner')
merged_ntd.head()
merged_ntd.to_csv('../Factors and Ridership Data/Tests/ntd_vrm.csv')
```

```python
### To take only the Bus and Rail modes

merged_ntd=merged_ntd.assign(RailBus=merged_ntd.apply(railbus,axis=1))
merged_ntd.head()
```

```python
merged_ntd=merged_ntd[(merged_ntd['RailBus']=='Bus')|(merged_ntd['RailBus']=='Rail')]
merged_ntd.head()
```

```python
### Remove rows which don't have either UPT or VRM for a month. Removes discripancy when aggregating over a year. 
##If a row has only VRM data, but no UPT data, when I aggregate over the year it will give low values for UPT corresponding to a high value of VRM

merged_ntd=merged_ntd[(merged_ntd['UPT'].notnull())|(merged_ntd['VRM'].notnull())]
```

```python
### Remove rows which dhas a 0 value for UPT or VRM for a month. Removes discripancy when aggregating over a year. 
##If a row has only VRM data, but no UPT data, when I aggregate over the year it will give low values for UPT corresponding to a high value of VRM

merged_ntd=merged_ntd[(merged_ntd['UPT']!=0)&(merged_ntd['VRM']!=0)]
```

```python
def extrapolate(x,y,z,a):
    rate1=((y-x)+(z-x)/2)/2
    rate2=((y-x)+(z-x)/2+(a-x)/3)/3
    temp=x+rate1
    if temp>0:
        new=x+rate1
    else:
        new=x+rate2
    return new

def extrapolate2(x,y,z,a):
    rate1=((x-y)+(x-z)/2)/2
    rate2=((x-y)+(x-z)/2+(x-a)/3)/3
    temp=x+rate1
    if temp>0:
        new=x+rate2
    else:
        new=x+rate1
    return new
```

```python
merge_2['2009'] = merge_2.apply(lambda x: extrapolate(x['2010'],x['2011'],x['2012'],
                                                      x['2013']), axis=1)
merge_2['2008'] = merge_2.apply(lambda x: extrapolate(x['2009'],x['2010'],x['2011'],x['2012'],
                                                      ), axis=1)
merge_2['2007'] = merge_2.apply(lambda x: extrapolate(x['2008'],x['2009'],
                                                      x['2010'],x['2011']), axis=1)
merge_2['2006'] = merge_2.apply(lambda x: extrapolate(x['2007'],x['2008'],x['2009'],
                                                      x['2010']), axis=1)
merge_2['2005'] = merge_2.apply(lambda x: extrapolate(x['2006'],x['2007'],
                                                      x['2008'],x['2009']), axis=1)
merge_2['2004'] = merge_2.apply(lambda x: extrapolate(x['2005'],x['2006'],x['2007'],
                                                      x['2008']), axis=1)
merge_2['2003'] = merge_2.apply(lambda x: extrapolate(x['2004'],x['2005'],
                                                      x['2006'],x['2007']), axis=1)
merge_2['2002'] = merge_2.apply(lambda x: extrapolate(x['2003'],x['2004'],x['2005'],
                                                      x['2006']), axis=1)
```
