```python
est=pd.read_csv('../Factors and Ridership Data/Model Estimation/Est10/tnc_revenue.csv')
est.drop(columns=['Unnamed: 0'],inplace=True)
est.head()
```

```python
est.MNAME.unique()
```

```python
cluster_1=['New York-Northern New Jersey-Long Island, NY-NJ-PA Metro Area']
cluster_2=['Boston-Cambridge-Quincy, MA-NH Metro Area','Chicago-Joliet-Naperville, IL-IN-WI Metro Area',
          'Los Angeles-Long Beach-Santa Ana, CA Metro Area','Miami-Fort Lauderdale-Pompano Beach, FL Metro Area',
          'Philadelphia-Camden-Wilmington, PA-NJ-DE-MD Metro Area','San Francisco-Oakland-Fremont, CA Metro Area',
          'Seattle-Tacoma-Bellevue, WA Metro Area','Washington-Arlington-Alexandria, DC-VA-MD-WV Metro Area']
cluster_3=['Baltimore-Towson, MD Metro Area','Dallas-Fort Worth-Arlington, TX Metro Area','Denver-Aurora-Broomfield, CO Metro Area',
          'Detroit-Warren-Livonia, MI Metro Area','Houston-Sugar Land-Baytown, TX Metro Area',
          'Minneapolis-St. Paul-Bloomington, MN-WI Metro Area','Phoenix-Mesa-Glendale, AZ Metro Area',
          'San Antonio-New Braunfels, TX Metro Area','San Diego-Carlsbad-San Marcos, CA Metro Area',
          'San Jose-Sunnyvale-Santa Clara, CA Metro Area','Minneapolis-St. Paul-Bloomington, MN-WI Metro Area']
def tnc_trip(df):   
    if df['Year']==2017:
        if df['MNAME'] in cluster_1:
            return 215
        elif df['MNAME'] in cluster_2:
            return 1009
        elif df['MNAME'] in cluster_3:
            return 171
        else:
            return 344
    else:
        return 0
    
def tnc_cluster(df):   
    if df['MNAME'] in cluster_1:
        return 'cluster_1'
    elif df['MNAME'] in cluster_2:
        return 'cluster_2'
    elif df['MNAME'] in cluster_3:
        return 'cluster_3'
    else:
        return 'cluster_4'
```

```python
est=est.assign(Cluster=est.apply(tnc_cluster,axis=1))
est.head()
#est=est.assign(TNC_TRIPS=est.apply(tnc_cluster,axis=1))
```

```python
est=est.assign(TNC_TRIP_2017_MILLION=est.apply(tnc_trip,axis=1))
est.head()
```

```python
est_2=est.drop_duplicates(['CBSA','MNAME','Year'])
est_2=est_2.drop(columns=['Mode'])
est_2.head()
```

```python
est_clusters=est_2.groupby(['Cluster','Year']).agg({'MNAME':'count','TNC_TRIP_2017_MILLION':'max'}).reset_index()
est_clusters=est_clusters.rename(columns={'MNAME':'CBSAs_WITH_TNC_in_Cluster'})
est_clusters
```

```python
est_clusters.sort_values(by=['Year','Cluster'])
```

```python
est_year=est_2.groupby(['Year','Cluster']).agg({'Tot_Pop':'sum','CBSA':'count'}).reset_index()
est_year=est_year.rename(columns={'CBSA':'ENTITIES'})
est_year
```

```python
est_year_grouped=est_year.groupby('Year').agg({'ENTITIES':'sum'})
est_year_grouped=est_year_grouped.rename(columns={'ENTITIES':'TOTAL_CBSAS_WITH_TNCS_YEARLY'})
est_year_2=pd.merge(est_year,est_year_grouped,on=['Year'])
est_year_2=est_year_2.rename(columns={'ENTITIES':'TOTAL_CBSAS_IN_CLUSTER_WITH_TNCS'})
est_year_2
```

```python
def cluster_trips(df):
    if df['Year']==2017:
        
        if df['Cluster']=='cluster_1':
            return 215
        elif df['Cluster']=='cluster_2':
            return 1009
        elif df['Cluster']=='cluster_3':
            return 171
        else:
            return 344
    else:
        return 'NA'
est_year_2['TRIPS']=est_year_2.apply(cluster_trips,axis=1)
#
est_year_2
```

```python
def uber_rev(df):
    if df['Year']==2012:
        return 15
    elif df['Year']==2013:
        return 103
    elif df['Year']==2014:
        return 445
    elif df['Year']==2015:
        return 1690
    elif df['Year']==2016:
        return 2964
    elif df['Year']==2017:
        return 7932
    else:
        return 11270

est_year_2['UBER_REVENUE']=est_year_2.apply(uber_rev,axis=1)
est_year_2
```

```python
def cluster_share(df):
    if df['Cluster']=='cluster_1':
        return 215/(215+1009+171+344)
    elif df['Cluster']=='cluster_2':
        return 1009/(215+1009+171+344)
    elif df['Cluster']=='cluster_3':
        return 171/(215+1009+171+344)
    else:
        return 344/(215+1009+171+344)
est_year_2['TRIPS_SHARE_2017']=est_year_2.apply(cluster_share,axis=1)
est_year_2
```

```python
est_year_2.sort_values(by=['Cluster','Year'])
```

```python
est_year_2['WEIGHTED_REV_SHARE']=est_year_2['UBER_REVENUE']*est_year_2['TOTAL_CBSAS_IN_CLUSTER_WITH_TNCS']/est_year_2['TOTAL_CBSAS_WITH_TNCS_YEARLY']
est_year_2
```

```python
est_year_3=est_year_2.groupby(['Year']).agg({'Tot_Pop':'sum','UBER_REVENUE':'max'})
est_year_3['UBER_REV_PER_CAPITA']=1000000*est_year_3['UBER_REVENUE']/est_year_3['Tot_Pop']
est_year_3=est_year_3.rename(columns={'Tot_Pop':'TOTAL_YEARLY_POPULATION','UBER_REVENUE':'TOTAL_YEARLY_REVENUE'})
est_year_3
```

```python
est_year_2=est_year_2.rename(columns={'Tot_Pop':'TOTAL_CLUSTER_POPULATION'})
est_year_2.drop(columns=['UBER_REVENUE'],inplace=True)
est_year_4=pd.merge(est_year_2,est_year_3,on=['Year'],how='left')
est_year_4
```

```python
est_year_4['TNC_REV_CLUSTER']=est_year_4['UBER_REV_PER_CAPITA']*est_year_4['TOTAL_CLUSTER_POPULATION']/1000000
est_year_4.head(10)
```

```python
def tnc_year(df):
    if df['Year']==0:
        if df['Tot_Pop']>30000000:
            x=2012
        elif df['Tot_Pop']>15000000:
            x=2013
        elif df['Tot_Pop']>5000000:
            x=2014
        elif df['Tot_Pop']>2000000:
            x=2015
        else:
            x=2016
    else:
        x=df['Year']
    return x
```
