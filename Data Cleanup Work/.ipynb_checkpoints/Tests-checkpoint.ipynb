```python
import pandas as pd
import numpy as np
```

```python
fare=pd.read_csv('../Factors and Ridership Data/Final Datasets/fare_yearly.csv')
```

```python
est_file=pd.read_csv('../Factors and Ridership Data/Estimation_File/estimation_file.csv')
```

```python
ridership=pd.read_csv('../Factors and Ridership Data/Final Datasets/annual_ridership_oct_24.csv',encoding='utf-8')
ridership.head()
```

```python
ridership=ridership[['CBSA','MNAME','RailBus','Year','UPT','VRM','UPT_ADJ','VRM_ADJ','UAPOPPCT']]
```

```python
est_file=est_file[['CBSA','Mode','MNAME','Year','UPT','VRM','UPT_ADJ','VRM_ADJ']]
```

```python
test_ridership=pd.merge(est_file,ridership,left_on=['CBSA','Mode','Year'],right_on=['CBSA','RailBus','Year'],how='left')
```
