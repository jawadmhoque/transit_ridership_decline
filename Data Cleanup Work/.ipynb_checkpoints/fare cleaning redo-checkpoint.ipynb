```python
rail=['AG','CC','CR','HR','LR','IP','MO','RL','RP','RS','SR','VT','MG']
bus=['MB','RB','CB','TB']
dr = ['DT','JT','VP','DR','PB']
other=['TR','FB']
def railbus(df):
    if df['MODE'] in rail:
        return 'Rail'
    elif df['MODE'] in bus:
        return 'Bus'
    elif df['MODE'] in dr:
        return 'DemandResponse'
    else:
        return "Other"
```

```python
def blank_ntd(df):
    if df['NTD_ID_y']==0:
        x=df['NTD_ID_x']
    elif df['NTD_ID_x']==0:
        x=df['NTD_ID_y']
    else: 
        x=df['NTD_ID_x']
    return x
```

```python
def blank_5_digit_id(df):
    if df['5_DIGIT_NTD_ID']=='Not Available':
        if len(str(df['NTD_ID']))==4:
            x=str(df['NTD_ID'])[0]+'0'+str(df['NTD_ID'])[2:]
        else:
            x=str(df['NTD_ID'])
    else:
        x=df['5_DIGIT_NTD_ID']
    return float(x)
```

```python
ntd_fare['5_Digit_NTD_ID']=ntd_fare.apply(blank_5_digit_id,axis=1)
ntd_fare.head()
```
