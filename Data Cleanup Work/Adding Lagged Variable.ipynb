```python
def lagged_column2(df,col,lag):
    x=np.where(df['min_year']!=df['Year'],df[col].shift(lag,fill_value=0),0)
    
    return x
```

```python
lagged_columns=[
 'VRM_ADJ',
 'FARE_per_UPT',
 'AVG_SPEED',
 'FARE_per_UPT_2018',
 'dockCt',
 'docklessCt',
 'scooterCt','TNC_FLAG']

for col in new_est.columns: 
    if col in lagged_columns:
        new_est[col+'_LAGGED_1'] = lagged_column2(new_est, col,1)
        new_est[col+'_LAGGED_2'] = lagged_column2(new_est, col,2)
        new_est[col+'_LAGGED_3'] = lagged_column2(new_est, col,3)
        
new_est['TNC_FLAG_LAGGED_4'] = new_est['TNC_FLAG'].shift(4).fillna(0)
new_est['TNC_FLAG_LAGGED_5'] = new_est['TNC_FLAG'].shift(5).fillna(0)
```

```python
new_est.to_csv('../Factors and Ridership Data/Estimation_File/estimation_file.csv')
```
