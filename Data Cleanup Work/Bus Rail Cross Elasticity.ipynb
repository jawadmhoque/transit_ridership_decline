```python
new_est=pd.merge(est_file,rail,on=['CBSA','Year'],suffixes=('','_COMPETING_RAIL'),how='left')
new_est.head()
```

```python
cross_elas=['UPT_ADJ','VRM_ADJ','VRH_ADJ','FARE_ADJ','FARE_per_UPT','AVG_SPEED']

bus_col=['UPT_ADJ_COMPETING_BUS',
 'VRM_ADJ_COMPETING_BUS',
 'VRH_ADJ_COMPETING_BUS',
 'FARE_ADJ_COMPETING_BUS',
 'FARE_per_UPT_COMPETING_BUS',
 'AVG_SPEED_COMPETING_BUS']
rail_col=['UPT_ADJ_COMPETING_RAIL',
 'VRM_ADJ_COMPETING_RAIL',
 'VRH_ADJ_COMPETING_RAIL',
 'FARE_ADJ_COMPETING_RAIL',
 'FARE_per_UPT_COMPETING_RAIL',
 'AVG_SPEED_COMPETING_RAIL']


if new_est2['Mode']=='Bus':
    new_est2['UPT_ADJ_COMPETING']=new_est2['UPT_ADJ_COMPETING_RAIL']
    new_est2['VRM_ADJ_COMPETING']=new_est2['VRM_ADJ_COMPETING_RAIL']
    new_est2['VRH_ADJ_COMPETING']=new_est2['VRH_ADJ_COMPETING_RAIL']
    new_est2['FARE_ADJ_COMPETING']=new_est2['FARE_ADJ_COMPETING_RAIL']
    new_est2['FARE_per_UPT_COMPETING']=new_est2['FARE_per_UPT_COMPETING_RAIL']
    new_est2['AVG_SPEED_COMPETING']=new_est2['AVG_SPEED_COMPETING_RAIL']
else:
    new_est2['UPT_ADJ_COMPETING']=new_est2['UPT_ADJ_COMPETING_BUS']
    new_est2['VRM_ADJ_COMPETING']=new_est2['VRM_ADJ_COMPETING_BUS']
    new_est2['VRH_ADJ_COMPETING']=new_est2['VRH_ADJ_COMPETING_BUS']
    new_est2['FARE_ADJ_COMPETING']=new_est2['FARE_ADJ_COMPETING_BUS']
    new_est2['FARE_per_UPT_COMPETING']=new_est2['FARE_per_UPT_COMPETING_BUS']
    new_est2['AVG_SPEED_COMPETING']=new_est2['AVG_SPEED_COMPETING_BUS']

new_est2.head()
```

```python
new_est2['UPT_ADJ_COMPETING']=np.where(new_est2['Mode']=='Bus',new_est2['UPT_ADJ_COMPETING_RAIL'],new_est2['UPT_ADJ_COMPETING_BUS'])
new_est2['VRM_ADJ_COMPETING']=np.where(new_est2['Mode']=='Bus',new_est2['VRM_ADJ_COMPETING_RAIL'],new_est2['VRM_ADJ_COMPETING_BUS'])
new_est2['VRH_ADJ_COMPETING']=np.where(new_est2['Mode']=='Bus',new_est2['VRH_ADJ_COMPETING_RAIL'],new_est2['VRH_ADJ_COMPETING_BUS'])
new_est2['FARE_ADJ_COMPETING']=np.where(new_est2['Mode']=='Bus',new_est2['FARE_ADJ_COMPETING_RAIL'],new_est2['FARE_ADJ_COMPETING_BUS'])
new_est2['FARE_per_UPT_COMPETING']=np.where(new_est2['Mode']=='Bus',new_est2['FARE_per_UPT_COMPETING_RAIL'],new_est2['FARE_per_UPT_COMPETING_BUS'])
new_est2['AVG_SPEED_COMPETING']=np.where(new_est2['Mode']=='Bus',new_est2['AVG_SPEED_COMPETING_RAIL'],new_est2['AVG_SPEED_COMPETING_BUS'])
new_est2.head()
```

```python
new_est2.drop(columns=['UPT_ADJ_COMPETING_BUS',
 'VRM_ADJ_COMPETING_BUS',
 'VRH_ADJ_COMPETING_BUS',
 'FARE_ADJ_COMPETING_BUS',
 'FARE_per_UPT_COMPETING_BUS',
 'AVG_SPEED_COMPETING_BUS','UPT_ADJ_COMPETING_RAIL',
 'VRM_ADJ_COMPETING_RAIL',
 'VRH_ADJ_COMPETING_RAIL',
 'FARE_ADJ_COMPETING_RAIL',
 'FARE_per_UPT_COMPETING_RAIL',
 'AVG_SPEED_COMPETING_RAIL'],inplace=True)
new_est2.head()
```

```python
def lagged_column2(df,col,lag):
    x=np.where(df['min_year']!=df['Year'],df[col].shift(lag),np.nan)
    
    return x
```

```python
lagged_columns=[
'UPT_ADJ_COMPETING',
 'VRM_ADJ_COMPETING',
 'VRH_ADJ_COMPETING',
 'FARE_ADJ_COMPETING',
 'FARE_per_UPT_COMPETING',
 'AVG_SPEED_COMPETING']

for col in new_est2.columns: 
    if col in lagged_columns:
        new_est2[col+'_LAGGED_1'] = lagged_column2(new_est2, col,1)
        new_est2[col+'_LAGGED_2'] = lagged_column2(new_est2, col,2)
        new_est2[col+'_LAGGED_3'] = lagged_column2(new_est2, col,3)
```
