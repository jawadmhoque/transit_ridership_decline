```python
def JTW(df):
    new_df=df.filter(['Id','Id2','Geography',
                      'Total; Estimate; Workers 16 years and over',
                       'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Car, truck, or van - Drove alone',
                     'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Car, truck, or van - Carpooled',
                     'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Public transportation (excluding taxicab)',
                      'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Walked',
                      'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Bicycle',
                      'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Taxicab, motorcycle, or other means',
                      'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Worked at home',
                     ],
                      axis=1)
    
    new_df2=new_df.rename(index=str, columns={
                      'Total; Estimate; Workers 16 years and over':'TOTAL_EMPLOYMENT',
                       'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Car, truck, or van - Drove alone':'JTW_DA_PCT',
                     'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Car, truck, or van - Carpooled':'JTW_CARPOOLED_PCT',
                     'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Public transportation (excluding taxicab)':'JTW_TRANSIT_PCT',
                      'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Walked':'JTW_WALK_PCT',
                      'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Bicycle':'JTW_BICYCLE_PCT',
                      'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Taxicab, motorcycle, or other means':'JTW_OTHER_PCT',
                      'Total; Estimate; Workers 16 years and over - MEANS OF TRANSPORTATION TO WORK - Worked at home':'JTW_HOME_PCT'
                     })

    return new_df2
```

```python
def JTW(df):
    new_df=df.filter(['Id','Id2','Geography',
                      'Car, truck, or van -- drove alone; Estimate; Workers 16 years and over',
                     'Car, truck, or van -- carpooled; Estimate; Workers 16 years and over',
                     'Public transportation (excluding taxicab); Estimate; Workers 16 years and over',
                     'Total; Estimate; Workers 16 years and over who did not work at home - TRAVEL TIME TO WORK - Mean travel time to work (minutes)'],
                      axis=1)
    
    new_df2=new_df.rename(index=str, columns={'Car, truck, or van -- drove alone; Estimate; Workers 16 years and over':'JTW_DA',
                     'Car, truck, or van -- carpooled; Estimate; Workers 16 years and over':'JTW_CARPOOL',
                     'Public transportation (excluding taxicab); Estimate; Workers 16 years and over':'JTW_TRANSIT',
                     'Total; Estimate; Workers 16 years and over who did not work at home - TRAVEL TIME TO WORK - Mean travel time to work (minutes)':'MEAN_TT'})

    return new_df2
```
