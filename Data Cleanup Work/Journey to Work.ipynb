```python
def JTW_acs2005(df):
    new_df=df.filter(['Id','Id2','Geography',
                      'Car, truck, or van -- drove alone; Estimate; Workers 16 years and over',
                     'Car, truck, or van -- carpooled; Estimate; Workers 16 years and over',
                     'Public transportation (excluding taxicab); Estimate; Workers 16 years and over',
                     'Mean travel time to work (minutes) for workers who did not work at home; Estimate; Workers 16 years and over'],
                      axis=1)
    
    new_df2=new_df.rename(index=str, columns={'Car, truck, or van -- drove alone; Estimate; Workers 16 years and over':'JTW_DA',
                     'Car, truck, or van -- carpooled; Estimate; Workers 16 years and over':'JTW_CARPOOL',
                     'Public transportation (excluding taxicab); Estimate; Workers 16 years and over':'JTW_TRANSIT',
                     'Mean travel time to work (minutes) for workers who did not work at home; Estimate; Workers 16 years and over':'MEAN_TT'})

    return new_df2
```

```python
def JTW(df):
    new_df=df.filter(['Id','Id2','Geography',
                      'Car, truck, or van -- drove alone; Estimate; Workers 16 years and over',
                     'Car, truck, or van -- carpooled; Estimate; Workers 16 years and over',
                     'Public transportation (excluding taxicab); Estimate; Workers 16 years and over',
                     'Total; Estimate; Workers 16 years and over who did not work at home - TRAVEL TIME TO WORK - Mean travel time to work (minutes)'],
                      axis=1)
    
    new_df2=new_df.rename(index=str, columns={'Car, truck, or van -- drove alone; Estimate; Workers 16 years and over':'JTW_DA',
                     'Car, truck, or van -- carpooled; Estimate; Workers 16 years and over':'JTW_CARPOOL',
                     'Public transportation (excluding taxicab); Estimate; Workers 16 years and over':'JTW_TRANSIT',
                     'Total; Estimate; Workers 16 years and over who did not work at home - TRAVEL TIME TO WORK - Mean travel time to work (minutes)':'MEAN_TT'})

    return new_df2
```

```python
def JTW_2010(df):
    new_df=df.filter(['Id','Id2','Geography',
                      'Car, truck, or van -- drove alone; Estimate; Workers 16 years and over',
                     'Car, truck, or van -- carpooled; Estimate; Workers 16 years and over',
                     'Public transportation (excluding taxicab); Estimate; Workers 16 years and over',
                     'Total; Estimate; TRAVEL TIME TO WORK - Mean travel time to work (minutes)'],
                      axis=1)

    new_df2=new_df.rename(index=str, columns={'Car, truck, or van -- drove alone; Estimate; Workers 16 years and over':'JTW_DA',
                     'Car, truck, or van -- carpooled; Estimate; Workers 16 years and over':'JTW_CARPOOL',
                     'Public transportation (excluding taxicab); Estimate; Workers 16 years and over':'JTW_TRANSIT',
                     'Total; Estimate; TRAVEL TIME TO WORK - Mean travel time to work (minutes)':'MEAN_TT'})

    return new_df2
```
