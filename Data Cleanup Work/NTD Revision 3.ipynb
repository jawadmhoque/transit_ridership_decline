```python
def lagged_column2(df,col,lag):
    x=np.where(df['min_year']!=df['Year'],df[col].shift(lag,fill_value=0),0)
   
    return x
def lagged_column3(df,col,lag):
    x=np.where(df['max_year']!=df['Year'],df[col].shift(lag,fill_value=0),0)
   
    return x

lagged_columns=['UPT/VRM','FARE/UPT','VRM/FARE']
 
for col in df_3_1.columns:
    if col in lagged_columns:
        df_3_1[col+'_Previous_Year'] = lagged_column2(df_3_1, col,1)
        df_3_1[col+'_Next_Year'] = lagged_column3(df_3_1, col,-1)
df_3_1.head()
```

```python
def upt_missing(df):
    if df['UPT']==0:
        if df['VRM']!=0:
            x=(df['UPT/VRM_Previous_Year']+df['UPT/VRM_Next_Year'])*df['VRM']/2
        elif df['FARE']!=0:
            x=df['FARE']/((df['FARE/UPT_Previous_Year']+df['FARE/UPT_Next_Year'])/2)
        else:
            x=0
    else:
        x=df['UPT']
    return x

def vrm_missing(df):
    if df['VRM']==0:
        if df['UPT']!=0:
            x=df['VRM']/((df['UPT/VRM_Previous_Year']+df['UPT/VRM_Next_Year'])/2)
        elif df['FARE']!=0:
            x=((df['VRM/FARE_Previous_Year']+df['VRM/FARE_Next_Year'])/2)*df['FARE']
        else:
            x=0
    else:
        x=df['VRM']
    return x

       
```

```python
def extrapolate(x,y,z,a):
    rate1=((y-x)+(z-x)/2)/2
    rate2=((y-x)+(z-x)/2+(a-x)/3)/3
    temp=x+rate1
    if temp>0:
        new=x+rate1
    else:
        new=x+rate2
    return new

def extrapolate2(x,y,z,a):
    rate1=((x-y)+(x-z)/2)/2
    rate2=((x-y)+(x-z)/2+(x-a)/3)/3
    temp=x+rate1
    if temp>0:
        new=x+rate2
    else:
        new=x+rate1
    return new
```
