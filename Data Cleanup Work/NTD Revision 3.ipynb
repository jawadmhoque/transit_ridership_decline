```python
import pandas as pd
import numpy as np
pd.options.display.max_columns=50
```

```python
### Read UPT and VRM data, from NTD March 2019
ntd_upt=pd.read_csv('../Factors and Ridership Data/Final Datasets/NTD_UPT.csv',encoding='utf-8')
#ntd_upt.head()
ntd_vrm=pd.read_csv('../Factors and Ridership Data/Final Datasets/NTD_VRM.csv',encoding='utf-8')
#ntd_vrm
```

```python
ntd_vrm.head()
```

```python
###Bus and Rail Specification

rail=['AG','CC','CR','HR','LR','IP','MO','RL','RP','RS','SR','VT','MG','YR']
bus=['MB','RB','CB','TB']
dr = ['DT','JT','VP','DR','PB']
other=['TR','FB']
def railbus(df):
    if df['Modes'] in rail:
        return 'Rail'
    elif df['Modes'] in bus:
        return 'Bus'
    elif df['Modes'] in dr:
        return 'DemandResponse'
    else:
        return "Other"
```

```python
### To take only the Bus and Rail modes

ntd_upt=ntd_upt.assign(RailBus=ntd_upt.apply(railbus,axis=1))
#ntd_vrm=

ntd_upt=ntd_upt[(ntd_upt['RailBus']=='Bus')|(ntd_upt['RailBus']=='Rail')]

#ntd_upt['VARIABLE']='UPT'
ntd_upt.head()
```

```python
ntd_vrm=ntd_vrm.assign(RailBus=ntd_vrm.apply(railbus,axis=1))


ntd_vrm=ntd_vrm[(ntd_vrm['RailBus']=='Bus')|(ntd_vrm['RailBus']=='Rail')]

#ntd_vrm['VARIABLE']='UPT'
ntd_vrm.head()
```

```python
grouped_upt=ntd_upt.groupby(by=['5 digit NTD ID','Agency','UZA Name','RailBus']).sum().reset_index()
```

```python
grouped_vrm=ntd_vrm.groupby(by=['5 digit NTD ID','Agency','UZA Name','RailBus']).sum().reset_index()
grouped_vrm.head()
```

```python
grouped_upt['VARIABLE']='UPT'
grouped_vrm['VARIABLE']='VRM'
```

```python
grouped_vrm.head()
```

```python
#grouped_ntd=
pd.concat([grouped_upt,grouped_vrm]).to_csv('../Factors and Ridership Data/Tests/concat_upt_vrm.csv')
```

```python
df_upt=pd.melt(ntd_upt, id_vars=list(ntd_upt)[0:9], value_vars=list(ntd_upt)[9:],var_name='MonthYear',
       value_name='UPT')
df_vrm=pd.melt(ntd_vrm, id_vars=list(ntd_vrm)[0:9], value_vars=list(ntd_vrm)[9:],var_name='MonthYear',
       value_name='VRM')
```

```python
df_upt['MonthYear']=df_upt['MonthYear'].apply(lambda x:x.replace('-',''))
df_upt['Month']=df_upt['MonthYear'].apply(lambda x:x[-3:].upper())
df_upt['Year']=df_upt['MonthYear'].apply(lambda x:str(x[:-3]))
df_upt.head()
```

```python
df_upt['Year']=df_upt['Year'].astype(int)
df_upt['Year']=df_upt['Year'].apply(lambda x: '0'+str(x) if x<10 else str(x))
df_upt['MonthYear']=df_upt['Month']+df_upt['Year']
df_upt.head()
```

```python
df_upt.drop(columns=['Month','Year'],inplace=True)
df_upt.head()
```

```python
### combined dataframe containing UPT and VRM values for each mode and each TOS at each Month

merged_ntd=pd.merge(df_upt,df_vrm,on=['5 digit NTD ID',
 'Agency',
 'Active',
 'Reporter Type',
 'UZA',
 'UZA Name',
 'Modes',
 'TOS',
 'MonthYear'],how='inner')

merged_ntd.drop(columns=['4 digit NTD ID_y'],inplace=True)
merged_ntd=merged_ntd.rename(columns={'4 digit NTD ID_x':'4 digit NTD ID'})

merged_ntd.head()
```

```python
merged_ntd['MonthText']=merged_ntd['MonthYear'].apply(lambda x:x[0:3]+'20'+x[3:])
merged_ntd.drop(columns=['MonthYear'],inplace=True)
merged_ntd['DateTime']=pd.to_datetime(merged_ntd['MonthText'])
merged_ntd.head()
```

```python
def missing_data(df):
    if (df['UPT'] is None)&(df['VRM'] is not None):
        note='ID '+str(df['5 digit NTD ID'])+' Mode '+df['Modes']+' TOS-'+df['TOS']+' UPT missing in Year '+str(df['Year'])+' month '+str(df['Month'])
        #flag=1
    elif (df['UPT'] is not None)&(df['VRM'] is None):
        note='ID '+str(df['5 digit NTD ID'])+' Mode '+df['Modes']+' TOS-'+df['TOS']+' VRM missing in Year '+str(df['Year'])+' month '+str(df['Month'])
        #flag=1
    elif (df['UPT'] is None)&(df['VRM'] is None):
        note='ID '+str(df['5 digit NTD ID'])+' Mode '+df['Modes']+' TOS-'+df['TOS']+' UPT and VRM missing in Year '+str(df['Year'])+' month '+str(df['Month'])
        #flag=1
    elif (df['UPT']==0)&(df['VRM']>0):
        note='ID '+str(df['5 digit NTD ID'])+' Mode '+df['Modes']+' TOS-'+df['TOS']+' UPT 0 value in Year '+str(df['Year'])+' month '+str(df['Month'])
        #flag=1
    elif (df['UPT']>0)&(df['VRM']==0):
        note='ID '+str(df['5 digit NTD ID'])+' Mode '+df['Modes']+' TOS-'+df['TOS']+' VRM 0 value in Year '+str(df['Year'])+' month '+str(df['Month'])
        #flag=1
    else:
        note='None'
    return note
    
```
