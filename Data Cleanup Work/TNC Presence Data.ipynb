```python
def tnc_year(df):
    if df['Year']==0:
        if df['Tot_Pop']>30000000:
            x=2012
        elif df['Tot_Pop']>15000000:
            x=2013
        elif df['Tot_Pop']>5000000:
            x=2014
        elif df['Tot_Pop']>2000000:
            x=2015
        else:
            x=2016
    else:
        x=df['Year']
    return x
```

```python
new['Year']=new['Year'].fillna(0)
new.head()
```

```python
#ntd_fare=fare_0517.assign(RailBus=fare_0517.apply(railbus,axis=1))
new=new.assign(TNC_ARRIVAL=new.apply(tnc_year,axis=1))
new.head()
```

```python
new=new.drop('Year',axis=1)
new.head()
```

```python
new.to_csv('../Factors and Ridership Data/Presence of TNCs/tnc_presence_data.csv')
```
