```python
df_3=pd.read_csv('../Factors and Ridership Data/Model Estimation/Est10/final_estimation_file_apta_added.csv',
                    encoding='utf-8')
df_3.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
list(df_3)
```

```python
df_3=df_3[['CBSA',
 'MNAME',
 'Mode',
 'Year','TNC_FLAG',
 'YEARS_SINCE_TNC']]
df_3.head()
```

```python
df_3=df_3[df_3['TNC_FLAG']==1]
df_3.head()
```

```python
df_4=pd.merge(df_3,df_2,on=['Year'])
df_4.head(20)
```

```python
df_4=df_4.sort_values(by=['CBSA','Mode','Year'])
df_4.head(20)
```

```python
df_4.head()
```

```python
#a=df_4.groupby(['CBSA','Mode']).min().reset_index()
earliest_year=df_4.groupby(['CBSA','Mode']).agg(min_year=('Year','min')).reset_index()
new_est=pd.merge(df_4,earliest_year,on=['CBSA','Mode'])
new_est.head()
```

```python
def lagged_column2(df,col,lag):
    x=np.where(df['min_year']!=df['Year'],df[col].shift(lag,fill_value=0),0)
    
    return x
```

```python
list(new_est)
```

```python
lagged_columns=[
'UBER_NET_REVENUE',
 'LYFT_TOTAL_RIDES']

for col in new_est.columns: 
    if col in lagged_columns:
        new_est[col+'_LAGGED_1'] = lagged_column2(new_est, col,1)

new_est.head()
```

```python
scale=new_est[new_est['min_year']==new_est['Year']]
scale=scale[['CBSA','MNAME','Mode','UBER_NET_REVENUE']]
scale=scale.rename(columns={'UBER_NET_REVENUE':'UBER_REV_FIRST_YEAR'})
scale.head()
```

```python
est=pd.merge(new_est,scale,on=['CBSA','MNAME','Mode'])
est.head()
```

```python
list(est)
```

```python
est=est[['CBSA',
 'MNAME',
 'Mode',
 'Year',
 'TNC_FLAG',
 'YEARS_SINCE_TNC','UBER_NET_REVENUE','UBER_REV_FIRST_YEAR']]
est.head()
```

```python
est['UBER_REV_SCALED']=est['UBER_NET_REVENUE']/est['UBER_REV_FIRST_YEAR']
est.drop(columns=['UBER_REV_FIRST_YEAR'],inplace=True)
est.head()
```

```python
est.to_csv('../Factors and Ridership Data/Model Estimation/Est10/tnc_revenue.csv')
```

```python
def tnc_year(df):
    if df['Year']==0:
        if df['Tot_Pop']>30000000:
            x=2012
        elif df['Tot_Pop']>15000000:
            x=2013
        elif df['Tot_Pop']>5000000:
            x=2014
        elif df['Tot_Pop']>2000000:
            x=2015
        else:
            x=2016
    else:
        x=df['Year']
    return x
```

```python
new['Year']=new['Year'].fillna(0)
new.head()
```

```python
#ntd_fare=fare_0517.assign(RailBus=fare_0517.apply(railbus,axis=1))
new=new.assign(TNC_ARRIVAL=new.apply(tnc_year,axis=1))
new.head()
```

```python
new=new.drop('Year',axis=1)
new.head()
```

```python
new.to_csv('../Factors and Ridership Data/Presence of TNCs/tnc_presence_data.csv')
```
