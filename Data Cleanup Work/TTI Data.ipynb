```python
import pandas as pd
import numpy as np
```

```python
tti=pd.read_csv('../Factors and Ridership Data/Congestion/TTI_Data.csv',
                    encoding='latin-1')
tti.head()
```

```python
cbsa=pd.read_csv('../Factors and Ridership Data/CBSA IDs/final_estimation_file_apta_added_greg.csv',
                    encoding='latin-1')
cbsa=cbsa[['CBSA','MNAME']]
cbsa=cbsa.drop_duplicates()
cbsa.head()
```

```python
cbsa['METRO']=cbsa.MNAME.apply(lambda x: x[:-11])
cbsa['METRO']=cbsa['METRO'].apply(lambda x: x.replace(', ',' '))
cbsa.head()
```

```python
tti_metro=tti[['UrbanArea','State']]
tti_metro=tti_metro.drop_duplicates()
tti_metro.head()
```

```python
cbsa_tti_names=pd.merge(cbsa,tti_metro,left_on=['METRO'],right_on=['UrbanArea'],how='outer',indicator=True)
cbsa_tti_names.to_csv('../Factors and Ridership Data/Congestion/TTI_Data_CBSA.csv')
```

```python
tti_data_metro=pd.read_csv('../Factors and Ridership Data/Congestion/TTI_Data_CBSA.csv')
tti_data_metro.head()
```

```python
final_tti=pd.merge(tti_data_metro,tti,
                  on=['UrbanArea','State'],
                  how='inner',indicator=True)
final_tti.head()
```

```python
final_tti=final_tti.rename(columns={' TTI ':'TTI',
       ' TotalDelay_hrs ':'TotalDelay_hrs', ' Delay_per_Auto_Commuter ':'Delay_per_Auto_Commuter'})
final_tti.columns
```

```python
test=final_tti[["CBSA", "MNAME",'UrbanArea','State','Year','TTI']]
test=test[test['Year']>2011]
test.head()
```

```python
final_tti.head()
```

```python
tot_del=final_tti[["CBSA", "MNAME",'UrbanArea','State','Year','TotalDelay_hrs']]
tot_del=tot_del[tot_del['Year']>2011]

del_per_commute=final_tti[["CBSA", "MNAME",'UrbanArea','State','Year','Delay_per_Auto_Commuter']]
del_per_commute=del_per_commute[del_per_commute['Year']>2011]

```

```python
tti_wide=test.pivot_table(index=["CBSA", "MNAME",'UrbanArea','State'], 
                    columns='Year', 
                    values='TTI').reset_index()
tti_wide[2018]=tti_wide[2017]+(tti_wide[2017]-tti_wide[2014])/4
tti_wide['TTI_INTERPOLATION']=np.where(tti_wide[2012].isnull(),1,0)

tti_wide[2013]=np.where(tti_wide[2013].isnull(),tti_wide[2014]-(tti_wide[2017]-tti_wide[2014])/4,tti_wide[2013])

tti_wide[2012]=np.where(tti_wide[2012].isnull(),tti_wide[2014]-2*(tti_wide[2017]-tti_wide[2014])/4,tti_wide[2012])




#tti_wide.drop(columns=['Year'],inplace=True)
tti_wide.head(10)
```

```python
tot_del['TotalDelay_hrs']=tot_del['TotalDelay_hrs'].apply(lambda x: int(x.replace(',','')))
```

```python
tot_delay_wide=tot_del.pivot_table(index=["CBSA", "MNAME",'UrbanArea','State'], 
                    columns='Year', 
                    values='TotalDelay_hrs').reset_index()
tot_delay_wide[2018]=tot_delay_wide[2017]+(tot_delay_wide[2017]-tot_delay_wide[2014])/4
tot_delay_wide['TOT_DEL_INTERPOLATION']=np.where(tot_delay_wide[2012].isnull(),1,0)

tot_delay_wide[2013]=np.where(tot_delay_wide[2013].isnull(),
                              tot_delay_wide[2014]-(tot_delay_wide[2017]-tot_delay_wide[2014])/4,
                              tot_delay_wide[2013])

tot_delay_wide[2012]=np.where(tot_delay_wide[2012].isnull(),
                              tot_delay_wide[2014]-2*(tot_delay_wide[2017]-tot_delay_wide[2014])/4,
                              tot_delay_wide[2012])




#tti_wide.drop(columns=['Year'],inplace=True)
tot_delay_wide.head(10)
```

```python
del_per_commute.head()
```

```python
del_per_auto_wide=del_per_commute.pivot_table(index=["CBSA", "MNAME",'UrbanArea','State'], 
                    columns='Year', 
                    values='Delay_per_Auto_Commuter').reset_index()
del_per_auto_wide[2018]=del_per_auto_wide[2017]+(del_per_auto_wide[2017]-del_per_auto_wide[2014])/4
del_per_auto_wide['DEL_PER_AUTO_INTERPOLATION']=np.where(del_per_auto_wide[2012].isnull(),1,0)

del_per_auto_wide[2013]=np.where(del_per_auto_wide[2013].isnull(),
                              del_per_auto_wide[2014]-(del_per_auto_wide[2017]-del_per_auto_wide[2014])/4,
                              del_per_auto_wide[2013])

del_per_auto_wide[2012]=np.where(del_per_auto_wide[2012].isnull(),
                              del_per_auto_wide[2014]-2*(del_per_auto_wide[2017]-del_per_auto_wide[2014])/4,
                              del_per_auto_wide[2012])

#tti_wide.drop(columns=['Year'],inplace=True)
del_per_auto_wide.head(10)
```

```python
tti_long=pd.melt(tti_wide,id_vars=['CBSA', 'MNAME', 'UrbanArea', 'State','TTI_INTERPOLATION'],
       value_vars=[2012, 2013, 2014, 2015, 2016,
       2017, 2018],value_name='TTI')
tti_long=tti_long.sort_values(by=['CBSA','MNAME','Year'])
tti_long
```

```python
del_per_auto_long=pd.melt(del_per_auto_wide,id_vars=['CBSA', 'MNAME', 'UrbanArea', 'State','DEL_PER_AUTO_INTERPOLATION'],
       value_vars=[2012, 2013, 2014, 2015, 2016,
       2017, 2018],value_name='DEL_PER_AUTO_COMMUTER')
del_per_auto_long=del_per_auto_long.sort_values(by=['CBSA','MNAME','Year'])
del_per_auto_long
```

```python
tot_delay_long=pd.melt(tot_delay_wide,id_vars=['CBSA', 'MNAME', 'UrbanArea', 'State'],
       value_vars=[2012, 2013, 2014, 2015, 2016,
       2017, 2018],value_name='TOT_DELAY_HRS')
tot_delay_long=tot_delay_long.sort_values(by=['CBSA','MNAME','Year'])
tot_delay_long
```

```python
tti_delay=pd.merge(tti_long,del_per_auto_long,on=['CBSA', 'MNAME', 'UrbanArea', 'State','Year'])

tti_delay_tot_del=pd.merge(tti_delay,tot_delay_long,on=['CBSA', 'MNAME', 'UrbanArea', 'State','Year'])

tti_delay_tot_del.drop(columns=['DEL_PER_AUTO_INTERPOLATION','UrbanArea','State'],inplace=True)

tti_delay_tot_del
```

```python
tti_delay_tot_del.to_csv('../Factors and Ridership Data/Congestion/CBSA_TTI_DELAY.csv')
```
