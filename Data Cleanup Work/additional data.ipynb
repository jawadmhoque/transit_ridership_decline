```python
import pandas as pd
import numpy as np
```

```python
pep_2018=pd.read_csv('PEP_2018.csv',encoding='utf-8')
cph=pd.read_excel('cph-t-2.xls',encoding='utf-8')
pep_2018.head()
```

```python
cph.head()
```

```python
cbsa=pd.read_csv('cbsa-2009.csv',encoding='latin-1')
cbsa.head()
```

```python
cbsa=cbsa[cbsa['LSAD']=='Metropolitan Statistical Area']
cbsa.head()
```

```python
cbsa['Geography']=cbsa['NAME'].apply(lambda x: x+" "+'Metro Area')
cbsa.head()
```

```python
population=pd.merge(cbsa,pep_2018,left_on='Geography',right_on='Geography')
population.head()
```

```python
pd.options.display.max_columns=50
```

```python
population.head()
```

```python
list(population)
```

```python
population=population.drop(columns=['LSAD','ESTIMATESBASE2000','POPESTIMATE2000'])
population.head()
```

```python
population=population.drop(columns=['NAME'])
population.head()
```

```python
list(population)
```

```python
population=population.reindex_axis(['CBSA','Geography','Jul-00','Jul-01','Jul-02',
                                    'Jul-03','Jul-04','Jul-05','Jul-06','Jul-07','Jul-08','Jul-09',
                                   'Apr-10','Jul-11','Jul-12','Jul-13',
                                    'Jul-14','Jul-15','Jul-16','Jul-17','Jul-18'],axis=1)
population.head()
```

```python
list(population)[2:22]
```

```python
population=pd.melt(population, id_vars=list(population)[0:2], value_vars=list(population)[2:],var_name='Month-Year',
       value_name='Population_Estimate')
population.head()
```

```python
population
```

```python
data=pd.to_datetime('Jul2000')
data
```

```python
population['MonthText']=population['Month-Year'].apply(lambda x: x[0:3]+'20'+x[4:])
population
```

```python
population['DateTime']=pd.to_datetime(population['MonthText'])
population
```

```python
population['Year']=population['MonthText'].apply(lambda x: pd.to_datetime(x).year)
population
```

```python
population=population.drop(columns=['Month-Year','MonthText'])
population.head()
```

```python
population=population.drop(columns=['CBSA'])
population.head()
```

```python
population.sort_values(by=['Geography','Year'])
```

```python
population.to_csv('annual-population.csv')
```
