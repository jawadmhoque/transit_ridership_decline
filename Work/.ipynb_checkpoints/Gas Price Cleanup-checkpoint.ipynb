```python
import pandas as pd
import numpy as np
```

```python
monthly_gasprice=pd.read_csv('../Factors and Ridership Data/Merged Data/cbsa_controls.csv',
                    encoding='latin-1')
monthly_gasprice
```

```python
monthly_gasprice=monthly_gasprice.dropna()
```

```python
annual_gasprice_average=monthly_gasprice.groupby(['CBSA_ID','cbsa_name','year']).mean()
```

```python
annual_gasprice_average.head()
```

```python
annual_gasprice_average=annual_gasprice_average.drop(columns=['month','popestimate','pop_interpolated','unemployment','employment','labor_force',
                             'unemployment_rate','employment_rate'])
```

```python
annual_gasprice_average.to_csv('../Factors and Ridership Data/Final Datasets/gasprice_yearly.csv')
```
