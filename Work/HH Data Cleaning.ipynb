```python
def HH_Inc(df):
    new_df=df.filter(['Id','Id2','Geography',
                      'Households; Estimate; Total - Less than $10,000',
                      'Households; Estimate; Total - $10,000 to $14,999',
                     'Households; Estimate; Total - $15,000 to $24,999',
                     'Households; Estimate; Total - $25,000 to $34,999',
                     'Households; Estimate; Total - $35,000 to $49,999',
                     'Households; Estimate; Total - $50,000 to $74,999',
                     'Households; Estimate; Total - $75,000 to $99,999',
                     'Households; Estimate; Median income (dollars)',
                     'Households; Estimate; Mean income (dollars)',
                     ],
                      axis=1)
    
    new_df['INC_U35']=new_df['Households; Estimate; Total - Less than $10,000']+new_df['Households; Estimate; Total - $10,000 to $14,999']+new_df['Households; Estimate; Total - $15,000 to $24,999']+new_df['Households; Estimate; Total - $25,000 to $34,999']
    new_df['INC_35_100']=new_df['Households; Estimate; Total - $35,000 to $49,999']+new_df['Households; Estimate; Total - $50,000 to $74,999']+new_df['Households; Estimate; Total - $75,000 to $99,999']
    new_df['INC_100P']=100-new_df['INC_U35']-new_df['INC_35_100']
    
    new_df2=new_df.rename(index=str, columns={'Households; Estimate; Median income (dollars)':'HH_MED_INC',
                                              'Households; Estimate; Mean income (dollars)':'HH_MEAN_INC'})
    
    new_df2=new_df2.drop(columns=['Households; Estimate; Total - Less than $10,000',
                                  'Households; Estimate; Total - $10,000 to $14,999',
                                  'Households; Estimate; Total - $15,000 to $24,999',
                                  'Households; Estimate; Total - $25,000 to $34,999',
                                  'Households; Estimate; Total - $35,000 to $49,999',
                                  'Households; Estimate; Total - $50,000 to $74,999',
                                  'Households; Estimate; Total - $75,000 to $99,999'])
    return new_df2
```
